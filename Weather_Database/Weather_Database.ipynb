```python
# Import the dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
# Create a set of 2,000 random latitudes and longitudes
lats = np.random.uniform(low=-90.000,high=90.000,size=2000)
lngs = np.random.uniform(low=-180.000,high=180.000,size=2000)
lat_lngs = zip(lats,lngs)
lat_lngs
```

```python
# Create a list of the latitudes and longitudes
coordinates = list(lat_lngs)

# Use the citipy module to determine city based on latitude and longitude
from citipy import citipy
```

```python
# Create a list for holding the cities
cities = []
# Identify the nearest city and country for each latitude and longitude combination
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    country = citipy.nearest_city(coordinate[0], coordinate[1]).country_code
    # If the city is unique, then we will add it to the cities list
    if city not in cities:
        cities.append(city)
# Print the city count to confirm sufficient count
len(cities)
```

```python
# Import the requests library
import requests

# Import the API key
from config import weather_api_key
```

```python
# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
```

```python
city_weather_df = pd.DataFrame()
for city in cities:
    city_url = url + "&q=" + city
    city_weather = requests.get(city_url)
    
    try:
        # Create a dictionary for this city and add it to the dataframe
        city_dict = {"City": city, "Country": city_weather.json()['sys']['country'], "Lat": city_weather.json()['coord']['lat'],"Lng": city_weather.json()['coord']['lon'], "Max Temp": city_weather.json()['main']['temp_max'], "Humidity": city_weather.json()['main']['humidity'],"Cloudiness": city_weather.json()['clouds']['all'], "Wind Speed": city_weather.json()['wind']['speed'], "Current Description": city_weather.json()['weather'][0]['description']}
        city_weather_df = city_weather_df.append(city_dict, ignore_index=True)
    
    except KeyError as sy:
        # Skip this city in the dataframe
        print(city + " data not found")

# Show part of dataframe, note there was one city with no data found
city_weather_df.head(10)
```

```python
# Export the dataframe to a csv file
city_weather_df.to_csv('WeatherPy_Database.csv', index=False)
```
